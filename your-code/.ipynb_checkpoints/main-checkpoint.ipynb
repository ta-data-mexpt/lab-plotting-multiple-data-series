## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
#agrupamos por el tipo de item y usamos .agg para sumar sus ventas por los 3 tipos 
total_sales = data.groupby('ItemType').agg({'RetailSales':'sum','RetailTransfers':'sum','WarehouseSales':'sum'})
total_sales.plot.bar()
#añadimos detalles a la gráfica 
plt.title("Sales by Item Type")
plt.xlabel('Item Type')
plt.ylabel('Sales')
plt.grid(True)
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
#agrupamos por el numero de proveedores que hay y usamos .agg para sumar sus ventas por los 3 tipos 
suppliers = data.groupby('Supplier').agg({'RetailSales':'sum','RetailTransfers':'sum','WarehouseSales':'sum'})
suppliers
```

```python
#agregamos una columna que es total_sales: va a sumar los 3 tipos de ventas de cada proveedor
suppliers['Total_Sales'] = suppliers['RetailSales'] + suppliers['RetailTransfers'] + suppliers['WarehouseSales']
suppliers
```

```python
#obtenemos al top10 de proveedores de acuerdo al total de sus ventas y los acomodamos de mayor a menor 
top_10 = suppliers.nlargest(columns= 'Total_Sales', n = 10).sort_values('Total_Sales', ascending=False)
top_10
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
#agrupamos por los meses que hay y usamos .agg para sacar la media de cada una de sus 3 tipos de venta 
Sales_per_month = data.groupby('Month').agg({'RetailSales':'mean','RetailTransfers':'mean','WarehouseSales':'mean'})
Sales_per_month
```

```python
#graficamos y damos detalles a la gráfica
Sales_per_month.plot()
plt.xticks(np.arange(1, 13, step=1))
plt.grid(True)
plt.title("Average sales by month")
plt.xlabel('Month')
plt.ylabel('Average Sales')
```

## 4. Plot the same information as above but as a bar chart.

```python
#graficamos lo mismo pero en barras y damos detalles a la gráfica
Sales_per_month.plot.bar()
plt.grid(True)
plt.title("Average sales by month")
plt.xlabel('Month')
plt.ylabel('Average Sales')
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
RetailSales_overTime = data.pivot_table(values = 'RetailSales', columns = 'ItemType', index = ['Year', 'Month'], aggfunc = sum)
RetailSales_overTime
```

```python
RetailSales_overTime.plot(figsize = (15,6))
```

## 6. Plot the same information as above but as a bar chart.

```python
RetailSales_overTime.plot.bar(figsize = (15,6))
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
#scatter
x = data['RetailSales']
y = data['RetailTransfers']
plt.scatter(x,y)
#podemos ver que la semana que vendió mas botellas es la semana 51
```

```python
#con hue=ItemType podemos diferenciar con colores los puntos
sns.lmplot(data = data, x = 'RetailSales', y = 'RetailTransfers', hue='ItemType', fit_reg=False)
```
