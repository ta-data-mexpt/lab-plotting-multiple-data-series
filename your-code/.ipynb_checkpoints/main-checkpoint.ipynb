# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import numpy as np

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

```python
data.info()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
data_to_group = data[["ItemType", "RetailSales", "RetailTransfers", "WarehouseSales"]]

grouped_data = data_to_group.groupby(data["ItemType"]).sum().reset_index()

melted_grouped = pd.melt(grouped_data, id_vars = "ItemType", var_name = "Sale Type", value_name = "Total Sales")

```

```python
sns.barplot(x= "ItemType", y = "Total Sales", hue = "Sale Type", data= melted_grouped)
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
data["Total Sales"] = data["RetailSales"] + data["WarehouseSales"]

grouped_supplier = data.groupby(data["Supplier"]).sum().reset_index().set_index("Supplier")
```

```python
sorted_grouped = grouped_supplier.sort_values(by = "Total Sales", ascending = False)

sorted_grouped.head()
```

```python
grouped_for_graph = sorted_grouped[["RetailSales", "WarehouseSales"]][0:10]
```

```python
grouped_for_graph.plot(kind='barh', stacked=True)
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
grouped_month = data.groupby("Month").mean().reset_index()
grouped_month.drop(columns = ["Year"], inplace = True)
```

```python
melted_month = pd.melt(grouped_month, id_vars = "Month", var_name = "Sale Type", value_name = "Total Sales")
```

```python
sns.lineplot(x='Month', y='Total Sales', hue='Sale Type', 
             data=melted_month)
```

## 4. Plot the same information as above but as a bar chart.

```python
sns.barplot(x='Month', y='Total Sales', hue='Sale Type', 
             data=melted_month)
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
grouped_month_year = data.groupby(["Month", "Year", "ItemType"]).sum().reset_index()
```

```python
grouped_month_year["time"] = grouped_month_year["Month"].astype(str) + "_" + grouped_month_year["Year"].astype(str)

grouped_month_year.drop(columns = ["RetailTransfers", "WarehouseSales", "Total Sales"], inplace = True)

grouped_month_year.time



```

```python
sorted_grouped_month_year = grouped_month_year.sort_values(by = ["Year", "Month"], ascending = True)

sorted_grouped_month_year.drop(columns = ["Year", "Month"], inplace = True)

sorted_grouped_month_year.head(20)
```

```python
sns.lineplot(x='time', y='RetailSales', hue='ItemType', 
             data=sorted_grouped_month_year)
```

## 6. Plot the same information as above but as a bar chart.

```python
sns.barplot(x='time', y='RetailSales', hue='ItemType', 
             data=sorted_grouped_month_year)
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
sns.lmplot(x="RetailSales", y="RetailTransfers", hue = "ItemType", fit_reg = False, data=data)
```
