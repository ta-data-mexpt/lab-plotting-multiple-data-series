# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import numpy as np

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
sales= data.groupby('ItemType').agg({'RetailSales':'sum', 'RetailTransfers':'sum', 'WarehouseSales': 'sum'}).reset_index()
```

```python
fig, ax = plt.subplots()
sales.plot.bar(x='ItemType', ax=ax)
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
supplier= data.groupby('Supplier')[['RetailSales', 'RetailTransfers', 'WarehouseSales']].sum().reset_index()
supplier['totalsales'] = supplier.sum(axis=1)
```

```python
t10= supplier.nlargest(10,'totalsales')
t10
```

```python
y_pos = np.arange(len(t10['Supplier']))
plt.barh(y_pos, t10['totalsales'])
plt.yticks(y_pos, t10['Supplier'])
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
sales_month= data.groupby('Month').agg({'RetailSales':'mean', 'RetailTransfers':'mean', 'WarehouseSales': 'mean'}).reset_index()
```

```python
sales_month
```

```python
x=sales_month['Month']

f, (ax1) = plt.subplots(1,figsize =(20,10))

plt.grid(True)
ax1.plot(x,sales_month['RetailSales'], label ='Retail Sales' )
ax1.plot(x, sales_month['RetailTransfers'], label = 'Retail Transfers' )
ax1.plot(x, sales_month['WarehouseSales'], label = 'Warehouse Sales')
ax1.legend()

plt.xlabel('Month')
```

## 4. Plot the same information as above but as a bar chart.

```python
fig, ax = plt.subplots(figsize=(20,10))
sales_month.plot.bar(x='Month', ax=ax)
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
data['day']=1
data['date']=pd.to_datetime(data[["Year", "Month", 'day']])
```

```python
sales= data.groupby(['date','ItemType', ]).agg({'RetailSales':'sum'}).reset_index()
sales
```

```python
x= sales['date']
items=sales['ItemType'].unique()

y1= sales[sales['ItemType']== 'BEER']
y2= sales[sales['ItemType']== 'LIQUOR']
y3= sales[sales['ItemType']== 'NON-ALCOHOL']
y4= sales[sales['ItemType']== 'REF']
y5= sales[sales['ItemType']== 'STR_SUPPLIES']
y6= sales[sales['ItemType']== 'WINE']
y7= sales[sales['ItemType']== 'DUNNAGE']
y8= sales[sales['ItemType']== 'KEGS']
```

```python

f, (ax1) = plt.subplots(1,figsize =(20,10))

plt.grid(True)

ax1.plot(y1['date'], y1['RetailSales'], label = 'Beer' )
ax1.plot(y2['date'], y2['RetailSales'], label = 'LIQUOR' )
ax1.plot(y3['date'], y3['RetailSales'], label = 'NON-ALCOHOL')
ax1.plot(y4['date'], y4['RetailSales'], label ='REF' )
ax1.plot(y5['date'], y5['RetailSales'], label = 'STR_SUPPLIES' )
ax1.plot(y6['date'], y6['RetailSales'], label = 'WINE')
ax1.plot(y7['date'], y7['RetailSales'], label ='DUNNAGE' )
ax1.plot(y8['date'], y8['RetailSales'], label = 'KEGS' )

ax1.legend()


```

## 6. Plot the same information as above but as a bar chart.

```python
ax=sns.barplot(x='date', y='RetailSales', hue='ItemType', data=sales)

ax.set_xticklabels(ax.get_xticklabels(), rotation = 90);
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
relationship= data.groupby('ItemType').agg({'RetailSales':'sum', 'RetailTransfers':'sum'}).reset_index()
relationship

```

```python
sns.set(context = 'paper', style = 'darkgrid', rc = {'figure.figsize': (10, 5)})

ax=sns.scatterplot(x='RetailSales', y='RetailTransfers', hue='ItemType', s= 200, legend="full", data=relationship)
```
