# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
grafica1 = data[['RetailSales','RetailTransfers','WarehouseSales','ItemType']].groupby(['ItemType']).agg('sum')
grafica1.plot.bar();
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
#Primero creamos la columna de TotalSales, que va a ser igual a RetailSales más WarehouseSales
data['TotalSales'] = (data['RetailSales'] + data['WarehouseSales'])

#Ponemos la información que queremos graficar
grafica2 = data[['Supplier', 'RetailSales','WarehouseSales', 'TotalSales']].groupby('Supplier').agg('sum').sort_values(by = ['TotalSales'], ascending = False)

#Ponemos el top10 para graficarlo
grafica2 = grafica2.head(10)

#Graficamos de manera horizontal
grafica2.plot.barh();
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
#Revisamos los años donde hay información
data['Year'].unique()
```

```python
#Revisamos los meses donde hay información
data['Month'].unique()
```

```python
#Revisamos los meses que tienen información de 2017
data2017 = data[data['Year'] == 2017]
data2017['Month'].unique()
```

```python
#Revisamos los meses que tienen información de 2018
data2018 = data[data['Year'] == 2018]
data2018['Month'].unique()
```

```python
#Creamos una nueva columna que contenga los datos de mes y año para poder comparar en el tiempo

#Primero tenemos que asignar los días de cada fecha. Para que todos sean iguales, pondremos el primero de cada mes
data['Day'] = 1

#Creamos la columna de Fecha
data['Fecha'] = pd.to_datetime(data[['Year', 'Month', 'Day']])
data.head()
```

```python
grafica3 = data[['Fecha','RetailSales','RetailTransfers','WarehouseSales']].groupby(['Fecha']).agg('mean')
grafica3.plot.line();
#grafica3.plot(x = 'Fecha', y = ['RetailTransfers','RetailSales','WarehouseSales']);
```

## 4. Plot the same information as above but as a bar chart.

```python
grafica4 = grafica3.copy()
grafica4.plot.bar();
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
#Con un groupby sale mal la gráfica.
#grafica5 = data[['Fecha','ItemType','RetailSales']].groupby(['ItemType','Fecha']).agg('sum').reset_index()

#Con un groupby sale más ordenado y graficamente correcto
grafica5 = data.pivot_table(values = 'RetailSales', index = ['Fecha'], columns = 'ItemType', aggfunc = 'sum')

grafica5.plot.line(figsize = (15,8));
```

## 6. Plot the same information as above but as a bar chart.

```python
grafica5.plot.bar(figsize = (15,8));
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
#Creamos el grid con un tamaño suficiente para que se vea correcto la gráfica
plt.figure(figsize = (15,8))

#Definimos las variables en los ejes correspondientes
x = data['RetailSales']
y = data['RetailTransfers']

#Seleccionamos los elementos a graficar y con el hue los agrupamos por tipo de item para que los ponga del mismo color
grafica7 = sns.scatterplot(x, y, hue = data['ItemType']);
grafica7
```
